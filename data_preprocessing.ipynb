```python
from google.colab import drive
import pandas as pd
file_location = '/content/drive/MyDrive/accelerometer.csv'
file_location1 = '/content/drive/MyDrive/gyroscope.csv'
accelerometer =pd.read_csv(file_location)
accelerometer.drop(columns=['time','repetition','uuid'], inplace=True)
gyroscope =pd.read_csv(file_location1)
gyroscope.drop(columns=['time','repetition','uuid'], inplace=True)
```

```python
accsitting= accelerometer[accelerometer['activity'] == 'CHAIR']
accstanding= accelerometer[accelerometer['activity'] == 'STANDING']
acclying= accelerometer[accelerometer['activity'] == 'LYING_UP']
acclying= acclying.append(accelerometer[accelerometer['activity'] == 'LYING_SIDE'].head(520))

gyrsitting= gyroscope[gyroscope['activity'] == 'CHAIR']
gyrstanding= gyroscope[gyroscope['activity'] == 'STANDING']
gyrlying= gyroscope[gyroscope['activity'] == 'LYING_UP']
```

```python
import matplotlib.pyplot as plt
plt.title("acc_x")
plt.plot(accsitting["acc_x"], label="sitting")
plt.plot(accstanding["acc_x"], label="standing")
plt.plot(acclying["acc_x"], label="lying")
plt.legend()
plt.show()

plt.title("acc_y")
plt.plot(accsitting["acc_y"], label="sitting")
plt.plot(accstanding["acc_y"], label="standing")
plt.plot(acclying["acc_y"], label="lying")
plt.legend()
plt.show()

plt.title("acc_z")
plt.plot(accsitting["acc_z"], label="sitting")
plt.plot(accstanding["acc_z"], label="standing")
plt.plot(acclying["acc_z"], label="lying")
plt.legend()
plt.show()
```

```python
plt.title("gyr_x")
plt.plot(gyrsitting["gyr_x"], label="sitting")
plt.plot(gyrstanding["gyr_x"], label="standing")
plt.plot(gyrlying["gyr_x"], label="lying")
plt.legend()
plt.show()

plt.title("gyr_x")
plt.plot(gyrsitting["gyr_y"], label="sitting")
plt.plot(gyrstanding["gyr_y"], label="standing")
plt.plot(gyrlying["gyr_y"], label="lying")
plt.legend()
plt.show()

plt.title("gyr_z")
plt.plot(gyrsitting["gyr_z"], label="sitting")
plt.plot(gyrstanding["gyr_z"], label="standing")
plt.plot(gyrlying["gyr_z"], label="lying")
plt.legend()
plt.show()
```
